```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("testset.csv")
```

```python
df.dropna(subset=[' _precipm'])
df.dropna(subset=[' _wdird'])
df.dropna(subset=[' _wdire'])
df.dropna(subset=[' _wgustm'])
df.dropna(subset=[' _heatindexm'])
```

```python
df['datetime_utc']=pd.to_datetime(df['datetime_utc'])
df['year']=df['datetime_utc'].dt.year
df['month']=df['datetime_utc'].dt.month
df.head()
```

```python
df[' _tempm'].describe()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['datetime_utc'],df[' _tempm'],color='purple')
plt.title("Temprature over time")
plt.xlabel("Year")
plt.ylabel("Temprature(C)")
plt.savefig("temprature.png")
plt.show()
```

```python
#zig-zag pattern every year indicates seasonal variation (summer=high temp,winter=low temp)
#peak shows hotter months, dips shows colder months
#unusually long spikes represents extreme heat conditions/data anamoly/sensor error/heat waves
#cycles appear consistent
```

```python
df[' _hum'].describe()
```

```python

plt.figure(figsize=(10,5))
plt.plot(df['datetime_utc'],df[' _hum'],color='purple')
plt.xlabel("Date")
plt.ylabel("Humidity(%)")
plt.title("Humidity over time")
plt.savefig("Humidity.png")
plt.show()
#0%-dry air 100%-air is fully saturated with moisture can cause rain or fog
```

```python
#Pattern suggests regular seosonal changes rather than a long term rise or drop in atmpospheric moisture
```

```python
df[' _pressurem'].describe()
```

```python
df = df[df[' _pressurem'] > 0]
df = df[df[' _pressurem'] < 2000]   
df[' _pressurem'].describe()
```

```python
df = df[(df[' _pressurem'] > 900) & (df[' _pressurem'] < 1100)]
df[' _pressurem'].describe()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['datetime_utc'],df[' _pressurem'],color='purple')
plt.xlabel("Date")
plt.ylabel("Pressure")
plt.title("Pressure over time")
plt.savefig("Pressure.png")
plt.show()
```

```python
#Shows zig-zag pattern due to continuos shifts between high and low pressure , indicating normal weather variations over time.
#unusually long spikes represents data anamoly/sensor error
```

```python
df = df[(df[ ' _wspdm']>0) & (df[ ' _wspdm'] <200)]
df[ ' _wspdm'].describe()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['datetime_utc'],df[ ' _wspdm'],color='purple')
plt.xlabel("Date")
plt.ylabel("wind Speed")
plt.title("Wind Speed over time")
plt.savefig("Wind.png")
plt.show()
```

```python
#Occasional extreme spikes represent- Severe storms /cyclones/Hurricane-like weather events/Sensor or data recording errors
#no long-term upward or downward trend, meaning wind intensity has not consistently increased or decreased over the years
#wind speeds may increase during monsoon or storm seasons, but stay calmer in summers/winters.
```
